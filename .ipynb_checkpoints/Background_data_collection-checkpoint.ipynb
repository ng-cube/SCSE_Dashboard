```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("professors.csv")
df
```

```python
def find_orcid(target_name, dblp_url):
    xml_url = dblp_url + ".xml"
    r = requests.get(xml_url)
    soup = BeautifulSoup(r.content, "xml")

    person_element = soup.find('person')
    if person_element:
        # Locate all the URL links within the person element
        url_links = person_element.find_all('url')
        orcid_url = None
        for link in url_links:
            if 'https://orcid.org/' in link.text:
                orcid_url = link.text
                break

        if orcid_url:
            print("ORCID URL for {}: {}".format(target_name, orcid_url))
            item = {
                "name":target_name,
                "orcid_URL":orcid_url
            }
            orcidURLs.append(orcid_url)
        else:
            print("ORCID URL not found for {}.".format(target_name))
            orcidURLs.append(np.nan)
    else:
        print("Person {} not found in the XML data.".format(target_name))
        orcidURLs.append(np.nan)
```

```python
orcidURLs=[]

for name,url in zip(df["Full Name"],df["DBLP URL"]):      
    if pd.isna(url):
        continue
    find_orcid(name, url)
```

```python
print(orcidURLs)
```

```python
df['orcid_url'] = orcidURLs
```
